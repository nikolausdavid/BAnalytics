```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from scipy.stats import f
from scipy.spatial.distance import pdist, squareform


class BankMarketingYClusters:
    def __init__(self, path, num_clusters=3, random_state=42):
        self.path = path
        self.num_clusters = num_clusters
        self.random_state = random_state
        self.df = None


        self.num_features = [
            "age", "balance", "duration",
            "campaign", "pdays", "previous"
        ]


        self.cat_features = [
            "job", "marital", "education",
            "default", "housing", "loan",
            "contact", "month", "poutcome"
        ]


        self.last_pair = None
        self.last_centers_pair_no_orig = None
        self.last_centers_pair_yes_orig = None
        self.last_means_no = None
        self.last_means_yes = None


    def load_data(self):
        df = pd.read_excel(self.path)
        # y in 0/1 umkodieren
        df["y"] = df["y"].map({"yes": 1, "no": 0})
        self.df = df
        print(f"Data loaded: {df.shape[0]} rows, {df.shape[1]} columns")
        return self


    def _spss_like_evaluation(self, df_sub, kmeans, used_features, label):
        """
            SPSS-like analysis:
            - Cluster sizes
            - Cluster centers (standardized, only used_features)
            - Means of all numeric variables (original scale)
            - ANOVA-like table for all numeric variables
            - Distances between cluster centers (standardized, used_features)
            - Total inertia
        """

        print(f"\n===== SPSS-like Evaluation for y={label} "
              f"(used: {used_features}) =====")


        cluster_sizes = df_sub["cluster"].value_counts().sort_index()
        print("\n--- Cluster Sizes ---")
        print(cluster_sizes)


        centers_scaled = pd.DataFrame(
            kmeans.cluster_centers_,
            columns=used_features
        )
        print("\n--- Cluster Centers (standardized, only used features) ---")
        print(centers_scaled.round(4))


        means_all = df_sub.groupby("cluster")[self.num_features].mean()
        print("\n--- Cluster Means (all numeric variables, original scale) ---")
        print(means_all.round(4))

        # --- ANOVA
        X_orig = df_sub[self.num_features].to_numpy()
        labels_cl = df_sub["cluster"].to_numpy()
        n = X_orig.shape[0]
        k = kmeans.n_clusters

        df_between = k - 1
        df_within = n - k

        rows = []
        for j, feat in enumerate(self.num_features):
            x = X_orig[:, j]
            grand_mean = x.mean()

            ss_total = np.sum((x - grand_mean) ** 2)

            ss_within = 0.0
            for c in range(k):
                mask = (labels_cl == c)
                x_c = x[mask]
                if len(x_c) > 1:
                    ss_within += np.sum((x_c - x_c.mean()) ** 2)

            ss_between = ss_total - ss_within
            ms_between = ss_between / df_between if df_between > 0 else np.nan
            ms_within = ss_within / df_within if df_within > 0 else np.nan

            if ms_within > 0 and not np.isnan(ms_within):
                F_stat = ms_between / ms_within
                p_value = 1 - f.cdf(F_stat, df_between, df_within)
            else:
                F_stat = np.nan
                p_value = np.nan

            rows.append({
                "Variable": feat,
                "SS Between": ss_between,
                "df Between": df_between,
                "MS Between": ms_between,
                "SS Within": ss_within,
                "df Within": df_within,
                "MS Within": ms_within,
                "F": F_stat,
                "p-value": p_value
            })

        anova_df = pd.DataFrame(rows)
        print("\n--- ANOVA-like table (all numeric variables) ---")
        print(anova_df.round(4))

        dist_matrix = squareform(pdist(kmeans.cluster_centers_, metric="euclidean"))
        dist_df = pd.DataFrame(
            dist_matrix,
            index=[f"Cluster {i}" for i in range(k)],
            columns=[f"Cluster {i}" for i in range(k)]
        )
        print("\n--- Distances between cluster centers "
              "(Euclidean, standardized, used features) ---")
        print(dist_df.round(4))

        inertia = kmeans.inertia_
        print("\n--- Total within-cluster sum of squares (Inertia) ---")
        print(inertia)

        if hasattr(self, "cat_features") and self.cat_features:
            print("\n=== Categorical variable profiles by cluster (row %) ===")
            for col in self.cat_features:
                if col in df_sub.columns:
                    ct = pd.crosstab(df_sub["cluster"], df_sub[col],
                                     normalize="index") * 100
                    print(f"\n>>> {col} (row % per cluster)")
                    print(ct.round(2))

        return {
            "cluster_sizes": cluster_sizes,
            "cluster_centers_standardized_used": centers_scaled,
            "cluster_means_all_numeric": means_all,
            "anova_table": anova_df,
            "center_distances": dist_df,
            "inertia": inertia
        }

    def _build_results_dataframe(self, df_sub, eval_dict, label):
        """
            Builds a long DataFrame with:
            - y
            - cluster
            - variable
            - category (for categorical variables)
            - mean (for numerical variables)
            - SS_between, SS_within, F, p (for numerical variables)
            - pct_in_cluster (for categorical variables)
            - cluster_size
            - inertia
        """

        records = []

        means = eval_dict["cluster_means_all_numeric"]
        anova = eval_dict["anova_table"].set_index("Variable")
        cluster_sizes = eval_dict["cluster_sizes"]
        inertia = eval_dict["inertia"]

        # --- Numerische Variablen ---
        for cluster_id in means.index:
            for var in self.num_features:
                records.append({
                    "y": label,
                    "cluster": cluster_id,
                    "variable": var,
                    "category": None,
                    "mean": means.loc[cluster_id, var],
                    "SS_between": anova.loc[var, "SS Between"],
                    "SS_within": anova.loc[var, "SS Within"],
                    "F": anova.loc[var, "F"],
                    "p": anova.loc[var, "p-value"],
                    "pct_in_cluster": None,
                    "cluster_size": cluster_sizes[cluster_id],
                    "inertia": inertia
                })


        for col in self.cat_features:
            if col in df_sub.columns:
                ct = (pd.crosstab(df_sub["cluster"], df_sub[col],
                                  normalize="index") * 100)

                for cluster_id in ct.index:
                    for cat in ct.columns:
                        records.append({
                            "y": label,
                            "cluster": cluster_id,
                            "variable": col,
                            "category": cat,
                            "mean": None,
                            "SS_between": None,
                            "SS_within": None,
                            "F": None,
                            "p": None,
                            "pct_in_cluster": ct.loc[cluster_id, cat],
                            "cluster_size": cluster_sizes[cluster_id],
                            "inertia": inertia
                        })

        return pd.DataFrame(records)


    def _plot_pair(self, df_sub, label, x_var, y_var, kmeans=None, scaler=None):
        """
        Scatterplot x_var vs. y_var, eingefärbt nach Cluster.
        Optional mit Clusterzentren im Originalraum.
        """
        plt.figure(figsize=(7, 6))
        sns.scatterplot(
            data=df_sub,
            x=x_var, y=y_var,
            hue="cluster",
            palette="viridis", s=50, alpha=0.8
        )
        plt.savefig(f"{label}_pair_plot")


        if kmeans is not None and scaler is not None:
            centers_scaled = kmeans.cluster_centers_
            centers_orig = scaler.inverse_transform(centers_scaled)
            plt.scatter(
                centers_orig[:, 0],
                centers_orig[:, 1],
                marker="X",
                s=200,
                edgecolor="black",
                linewidth=1.5,
                alpha=0.9,
                label="Cluster centers"
            )
            plt.legend()

        plt.title(f"y={label} | {x_var} vs. {y_var} (K-Means clusters)")
        plt.tight_layout()
        plt.show()


    def run_for_pair(self, x_var="duration", y_var="campaign", num_clusters=None):
        """
        Führt für ein gewähltes Feature-Paar (x_var, y_var):
        - K-Means-Clustering getrennt für y=0 und y=1
        - SPSS-ähnliche Kennzahlen (inkl. aller numerischen Variablen)
        - kategoriale Profile
        - je einen Plot (y=0, y=1)

        Rückgabe:
        - df_no: Daten mit y=0 + Clusterlabel
        - df_yes: Daten mit y=1 + Clusterlabel
        - res_no: großer Ergebnis-DataFrame für y=0
        - res_yes: großer Ergebnis-DataFrame für y=1
        """

        if self.df is None:
            self.load_data()

        if num_clusters is None:
            num_clusters = self.num_clusters

        if x_var not in self.num_features or y_var not in self.num_features:
            raise ValueError("x_var and y_var have to be contained in self.num_features.")

        print(f"\n=============================")
        print(f"Running K-Means on pair: {x_var} vs. {y_var}")
        print(f"Clusters: {num_clusters}")
        print(f"=============================\n")


        self.last_pair = (x_var, y_var)


        df_no = self.df[self.df["y"] == 0].copy()
        df_yes = self.df[self.df["y"] == 1].copy()


        print("\n### Clustering for y=0 ###")
        scaler_no = StandardScaler().fit(df_no[[x_var, y_var]])
        X_no_scaled = scaler_no.transform(df_no[[x_var, y_var]])

        kmeans_no = KMeans(
            n_clusters=num_clusters,
            random_state=self.random_state,
            n_init="auto"
        )
        df_no["cluster"] = kmeans_no.fit_predict(X_no_scaled)

        eval_no = self._spss_like_evaluation(
            df_sub=df_no,
            kmeans=kmeans_no,
            used_features=[x_var, y_var],
            label=0
        )

        self._plot_pair(df_no, label=0, x_var=x_var, y_var=y_var,
                        kmeans=kmeans_no, scaler=scaler_no)

        res_no = self._build_results_dataframe(df_no, eval_no, label=0)


        centers_no_pair_orig = scaler_no.inverse_transform(kmeans_no.cluster_centers_)
        self.last_centers_pair_no_orig = pd.DataFrame(
            centers_no_pair_orig,
            columns=[x_var, y_var]
        )
        self.last_means_no = eval_no["cluster_means_all_numeric"]

        # === y=1 ===
        print("\n### Clustering for y=1 ###")
        scaler_yes = StandardScaler().fit(df_yes[[x_var, y_var]])
        X_yes_scaled = scaler_yes.transform(df_yes[[x_var, y_var]])

        kmeans_yes = KMeans(
            n_clusters=num_clusters,
            random_state=self.random_state,
            n_init="auto"
        )
        df_yes["cluster"] = kmeans_yes.fit_predict(X_yes_scaled)

        eval_yes = self._spss_like_evaluation(
            df_sub=df_yes,
            kmeans=kmeans_yes,
            used_features=[x_var, y_var],
            label=1
        )

        self._plot_pair(df_yes, label=1, x_var=x_var, y_var=y_var,
                        kmeans=kmeans_yes, scaler=scaler_yes)

        res_yes = self._build_results_dataframe(df_yes, eval_yes, label=1)


        centers_yes_pair_orig = scaler_yes.inverse_transform(kmeans_yes.cluster_centers_)
        self.last_centers_pair_yes_orig = pd.DataFrame(
            centers_yes_pair_orig,
            columns=[x_var, y_var]
        )
        self.last_means_yes = eval_yes["cluster_means_all_numeric"]

        return df_no, df_yes, res_no, res_yes


    def get_cluster_centers_original_scale(self, y=0, mode="pair"):
        """
            Returns cluster centers on the original scale based on the last run_for_pair call.
            
            Parameters:
            - y: 0 or 1 (Non-Subscriber / Subscriber)
            - mode:
                - "pair": centers in the last used 2D pair (e.g., duration/campaign)
                - "all":  means of all numerical variables per cluster
            
            Return:
            - DataFrame with clusters as index and variables as columns
        """
        if self.last_pair is None:
            raise RuntimeError("run_for_pair() wasn't executed yet.")

        if y not in (0, 1):
            raise ValueError("y hast to be 0 or 1.")

        if mode not in ("pair", "all"):
            raise ValueError("mode hast to be 'pair' or 'all'.")

        if mode == "pair":
            if y == 0:
                if self.last_centers_pair_no_orig is None:
                    raise RuntimeError("No saved centers for y=0.")
                return self.last_centers_pair_no_orig.copy()
            else:
                if self.last_centers_pair_yes_orig is None:
                    raise RuntimeError("No saved centers for y=1.")
                return self.last_centers_pair_yes_orig.copy()

        # mode == "all"
        if y == 0:
            if self.last_means_no is None:
                raise RuntimeError("No saved means für y=0.")
            return self.last_means_no.copy()
        else:
            if self.last_means_yes is None:
                raise RuntimeError("No saved means for y=1.")
            return self.last_means_yes.copy()
```

```python
bm = BankMarketingYClusters("Bank Marketing Data.xlsx", num_clusters=3).load_data()

df_no, df_yes, res_no, res_yes = bm.run_for_pair("duration", "campaign")

# 3) Cluster-"Centres" (Means) of all numeric variables, y=0:
centers_no_all = bm.get_cluster_centers_original_scale(y=0, mode="all")
centers_no_pair = bm.get_cluster_centers_original_scale(y=0, mode="pair")

# 4) Cluster-"Centres" (Means) of all numeric variables, y=1:
centers_yes_all = bm.get_cluster_centers_original_scale(y=1, mode="all")
centers_yes_pair = bm.get_cluster_centers_original_scale(y=1, mode="pair")
```
