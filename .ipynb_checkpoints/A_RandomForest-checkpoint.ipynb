```python
import numpy as np
import pandas as pd
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

import shap
import matplotlib.pyplot as plt
from types import MethodType
import pandas as pd
import numpy as np

RANDOM_STATE = 42
TEST_SIZE = 0.30
TARGET = "y_bin"


class BankMarketingModelHybrid:
    def __init__(self, path,
                 drop_duration=True,
                 test_size=TEST_SIZE,
                 random_state=RANDOM_STATE,
                 sampling_mode="hybrid",
                 oversample_ratio=0.7,
                 target_ratio=1.0):
        self.path = Path(path)
        self.drop_duration = drop_duration
        self.test_size = test_size
        self.random_state = random_state
        self.sampling_mode = sampling_mode
        self.oversample_ratio = oversample_ratio
        self.target_ratio = target_ratio

        self.df = None
        self.cohorts = {}
        self.models = {}
        self.metas = {}
        self.splits = {}

    def load_and_clean(self):
        df = pd.read_excel(self.path, sheet_name=0)
        df.columns = [str(c).strip().lower() for c in df.columns]
        if "y" not in df.columns:
            raise ValueError("Target column 'y' not found.")
        df["y"] = df["y"].astype(str).str.lower().str.strip()
        df[TARGET] = (df["y"] == "yes").astype(int)
        for c in df.select_dtypes(include=["object"]).columns:
            df[c] = df[c].astype(str).str.lower().str.strip()
        self.df = df
        print("Loaded shape:", df.shape)

    def sanity_and_split(self):
        df = self.df.copy()
        print("Inconsistent rows (pdays=-1 & previous!=0):", len(df[(df["pdays"] == -1) & (df["previous"] != 0)]))
        first_time = df[(df["pdays"] == -1) & (df["previous"] == 0)].copy()
        previously_contacted = df[(df["pdays"] != -1) | (df["previous"] > 0)].copy()
        self.cohorts["first_time"] = first_time
        self.cohorts["previously_contacted"] = previously_contacted
        print("First-time:", first_time.shape, "Previously-contacted:", previously_contacted.shape)

    def _build_pipeline(self, df_in):
        drop_cols = [TARGET, "y"]
        if self.drop_duration and "duration" in df_in.columns:
            drop_cols.append("duration")
        X = df_in.drop(columns=[c for c in drop_cols if c in df_in.columns])
        y = df_in[TARGET].astype(int)

        cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
        num_cols = X.select_dtypes(include=[np.number, "bool"]).columns.tolist()

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=self.test_size, stratify=y, random_state=self.random_state
        )

        pre = ColumnTransformer(
            transformers=[
                ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=True), cat_cols),
                ("num", "passthrough", num_cols),
            ],
            sparse_threshold=1.0,
            remainder="drop",
        )

        to_dense = FunctionTransformer(
            lambda X: X.toarray() if hasattr(X, "toarray") else np.asarray(X),
            accept_sparse=True, validate=False
        )

        rf = RandomForestClassifier(
            n_estimators=300,
            max_depth=None,
            min_samples_split=4,
            min_samples_leaf=2,
            random_state=self.random_state,
            n_jobs=-1,
            class_weight=None
        )

        if self.sampling_mode == "hybrid":
            pipe = ImbPipeline([
                ("pre", pre),
                ("to_dense", to_dense),
                ("ros", RandomOverSampler(sampling_strategy=self.oversample_ratio, random_state=self.random_state)),
                ("rus", RandomUnderSampler(sampling_strategy=self.target_ratio, random_state=self.random_state)),
                ("clf", rf),
            ])
        else:
            from sklearn.pipeline import Pipeline as SkPipeline
            rf.set_params(class_weight="balanced_subsample")
            pipe = SkPipeline([("pre", pre), ("clf", rf)])

        meta = {"cat_cols": cat_cols, "num_cols": num_cols}
        return pipe, meta, (X_train, X_test, y_train, y_test)

    def _feature_names(self, pipe, meta):
        pre = pipe.named_steps["pre"]
        ohe = pre.named_transformers_["cat"]
        cat_names = ohe.get_feature_names_out(meta["cat_cols"]).tolist()
        names = cat_names + meta["num_cols"]
        return names

    def fit_evaluate(self, cohort_key):
        dfc = self.cohorts[cohort_key]
        pipe, meta, splits = self._build_pipeline(dfc)
        Xtr, Xte, ytr, yte = splits

        pipe.fit(Xtr, ytr)
        y_pred = pipe.predict(Xte)
        y_proba = pipe.predict_proba(Xte)[:, 1]
        auc = roc_auc_score(yte, y_proba)
        print(f"[{cohort_key}] ROC-AUC (test): {auc:.3f}")

        report = pd.DataFrame(classification_report(yte, y_pred, output_dict=True)).T.round(3)
        cm = pd.DataFrame(confusion_matrix(yte, y_pred), index=["Actual 0", "Actual 1"], columns=["Pred 0", "Pred 1"])
        print("REPORT")
        display(report)
        print("CM")
        display(cm)

        fpr, tpr, _ = roc_curve(yte, y_proba)
        plt.figure()
        plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
        plt.plot([0, 1], [0, 1], linestyle="--")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"ROC Curve — {cohort_key}")
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.savefig(f"{cohort_key}_ROC_CURVE")
        plt.show()

        self.models[cohort_key] = pipe
        self.metas[cohort_key] = meta
        self.splits[cohort_key] = splits

    def plot_feature_importances(self, cohort_key, top_k=15):
        pipe = self.models[cohort_key]
        meta = self.metas[cohort_key]
        names = self._feature_names(pipe, meta)
        importances = pd.Series(pipe.named_steps["clf"].feature_importances_, index=names).sort_values(ascending=False)
        top = importances.head(top_k)
        display(top.to_frame("importance"))
        plt.figure()
        top.iloc[::-1].plot(kind="barh")
        plt.xlabel("Importance")
        plt.title(f"Top {top_k} Feature Importances — {cohort_key}")
        plt.tight_layout()
        plt.show()

    def shap_values(self, cohort_key, max_rows=400, summary=True, bar=True, force=False):
        import numpy as np
        import shap
        import matplotlib.pyplot as plt
    
        # SHAP compatibility for deprecated numpy aliases
        if not hasattr(np, "bool"): np.bool = np.bool_
        if not hasattr(np, "str"):  np.str  = str
    
        pipe  = self.models[cohort_key]
        meta  = self.metas[cohort_key]
        Xtr, Xte, ytr, yte = self.splits[cohort_key]
    
        # Subsample test set for speed
        Xs = Xte.sample(min(len(Xte), max_rows), random_state=self.random_state)
    
        # Use the SAME preprocessor the model was trained with
        pre   = pipe.named_steps["pre"]
        Xtrns = pre.transform(Xs)
        X_mat = Xtrns.toarray() if hasattr(Xtrns, "toarray") else np.asarray(Xtrns)
    
        rf = pipe.named_steps["clf"]
        explainer = shap.TreeExplainer(rf)
    
        vals = explainer.shap_values(X_mat)
    
        # ---- Normalize SHAP outputs across SHAP versions / models ----
        if isinstance(vals, list):
            # classic API: list per class
            sv = vals[1] if len(vals) > 1 else vals[0]
            ev = explainer.expected_value
            base = ev[1] if (hasattr(ev, "__len__") and len(ev) > 1) else ev
        else:
            # new API: single array (n_samples, n_features) or (n_samples, n_features, n_outputs)
            if vals.ndim == 3:
                sv = vals[..., 1] if vals.shape[-1] > 1 else vals[..., 0]
            else:
                sv = vals
            ev = explainer.expected_value
            base = ev[1] if (hasattr(ev, "__len__") and len(ev) > 1) else ev
    
        # Final safety check
        assert sv.shape[:2] == X_mat.shape[:2], \
            f"SHAP shape {sv.shape} vs X {X_mat.shape} — versions disagree."
    
        names = self._feature_names(pipe, meta)
    
        if summary:
            plt.title(f"SHAP Summary — {cohort_key}")
            shap.summary_plot(sv, features=X_mat, feature_names=names, show=False)
            plt.tight_layout()
            plt.savefig(f"SHAP_Plot_{cohort_key}")
            plt.show()
    
        if bar:
            plt.title(f"Mean |SHAP| — {cohort_key}")
            shap.summary_plot(sv, features=X_mat, feature_names=names, plot_type="bar", show=False)
            plt.tight_layout(); plt.show()
    
        if force:
            idx = np.random.randint(0, sv.shape[0])
            shap.force_plot(base, sv[idx, :], matplotlib=True)
            plt.title(f"SHAP Force Plot — {cohort_key}")
            plt.savefig(f"SHAP_{cohort_key}.png")
            plt.show()
    
    def get_encoded_data(self, cohort_key, which="train", resampled=False, return_y=True, max_rows=None):
        """
        Returns the encoded feature matrix as a pandas DataFrame (and y as a Series if return_y=True).
        
        Parameters
        ----------
        cohort_key : str
            "first_time" or "previously_contacted"
        which : str
            "train", "test", or "all" (all = concatenate train and test in that order)
        resampled : bool
            If True, returns the *resampled* training set (oversampling + undersampling) — only valid for which="train".
            Ignored for which="test"/"all".
        return_y : bool
            If True, also return the target as a Series.
        max_rows : int or None
            Optional row cap (after transformation) to limit size for display/performance.
        """
        if cohort_key not in self.models:
            raise ValueError(f"Model for cohort '{cohort_key}' not fitted yet. Call fit_evaluate('{cohort_key}') first.")
    
        pipe = self.models[cohort_key]
        meta = self.metas[cohort_key]
        Xtr, Xte, ytr, yte = self.splits[cohort_key]
    
        # helper to encode a matrix and return DataFrame with names
        def encode_to_df(X):
            pre = pipe.named_steps["pre"]
            Xt = pre.transform(X)
            if hasattr(Xt, "toarray"):
                Xt = Xt.toarray()
            else:
                Xt = np.asarray(Xt)
            names = self._feature_names(pipe, meta)
            return pd.DataFrame(Xt, columns=names)
    
        if which not in {"train", "test", "all"}:
            raise ValueError("Argument 'which' must be 'train', 'test', or 'all'.")
    
        if resampled and which != "train":
            print("Note: resampled=True is only applicable for which='train'. Ignoring for this request.")
            resampled = False
    
        if resampled:
            # Ensure we have samplers in pipeline
            step_names = [name for name, _ in pipe.steps]
            if "ros" in step_names or "rus" in step_names:
                # Fit-resample through all steps up to (but not including) the classifier
                X_res, y_res = pipe[:-1].fit_resample(Xtr, ytr)
                # Feature names are from pre step; X_res is already dense after to_dense
                names = self._feature_names(pipe, meta)
                X_df = pd.DataFrame(X_res, columns=names)
                y_ser = pd.Series(y_res, name="y_bin")
                if max_rows is not None:
                    X_df = X_df.head(max_rows)
                    y_ser = y_ser.head(max_rows)
                return (X_df, y_ser) if return_y else X_df
            else:
                raise ValueError("Resampling is not enabled in the current pipeline (sampling_mode='none').")
    
        # Non-resampled paths
        if which == "train":
            X_df = encode_to_df(Xtr)
            y_ser = ytr.reset_index(drop=True)
            if max_rows is not None:
                X_df = X_df.head(max_rows)
                y_ser = y_ser.head(max_rows)
            return (X_df, y_ser) if return_y else X_df
    
        if which == "test":
            X_df = encode_to_df(Xte)
            y_ser = yte.reset_index(drop=True)
            if max_rows is not None:
                X_df = X_df.head(max_rows)
                y_ser = y_ser.head(max_rows)
            return (X_df, y_ser) if return_y else X_df
    
        # which == "all"
        Xtr_df = encode_to_df(Xtr)
        Xte_df = encode_to_df(Xte)
        X_df = pd.concat([Xtr_df, Xte_df], axis=0, ignore_index=True)
        if return_y:
            y_ser = pd.concat([ytr.reset_index(drop=True), yte.reset_index(drop=True)], axis=0, ignore_index=True)
            if max_rows is not None:
                X_df = X_df.head(max_rows)
                y_ser = y_ser.head(max_rows)
            return X_df, y_ser
        else:
            if max_rows is not None:
                X_df = X_df.head(max_rows)
            return X_df


    def plot_sample_tree(self, cohort_key, estimator_idx=0, max_depth=3, dpi=150, figsize=(20, 10), save_path=None):
        """
        Plot a single decision tree from the fitted RandomForest for the given cohort.
    
        Parameters
        ----------
        cohort_key : str
            "first_time" or "previously_contacted"
        estimator_idx : int
            Index of the tree inside the forest (0..n_estimators-1)
        max_depth : int or None
            Limit depth for readability; None = full tree
        dpi : int
            Figure DPI
        figsize : tuple
            Matplotlib figure size
        save_path : str or None
            If provided, saves the plot to this path (e.g. 'tree_first_time.png')
        """
        import matplotlib.pyplot as plt
        from sklearn.tree import plot_tree
    
        if cohort_key not in self.models:
            raise ValueError(f"Model for cohort '{cohort_key}' not fitted yet. Call fit_evaluate('{cohort_key}') first.")
    
        pipe = self.models[cohort_key]
        meta = self.metas[cohort_key]
    
        # Grab a single decision tree from the RF
        rf = pipe.named_steps["clf"]
        if estimator_idx < 0 or estimator_idx >= len(rf.estimators_):
            raise IndexError(f"estimator_idx out of range: 0..{len(rf.estimators_)-1}")
    
        tree = rf.estimators_[estimator_idx]
    
        # Build proper feature names from the OHE + numeric passthrough
        pre = pipe.named_steps["pre"]
        ohe = pre.named_transformers_["cat"]
        cat_names = ohe.get_feature_names_out(meta["cat_cols"]).tolist()
        feat_names = cat_names + meta["num_cols"]
    
        plt.figure(figsize=figsize, dpi=dpi)
        plot_tree(
            tree,
            feature_names=feat_names,
            filled=True,
            rounded=True,
            max_depth=max_depth,
            fontsize=8
        )
        plt.title(f"RandomForest – Sample Tree (cohort='{cohort_key}', idx={estimator_idx}, max_depth={max_depth})")
        if save_path:
            plt.tight_layout()
            plt.savefig(save_path, dpi=dpi, bbox_inches="tight")
        plt.show()
    
```

```python
# Run with hybrid sampling
bm2 = BankMarketingModelHybrid("Bank Marketing Data.xlsx",
                               drop_duration=True,
                               sampling_mode="hybrid",
                               oversample_ratio=0.7,
                               target_ratio=1.0)

bm2.load_and_clean()
bm2.sanity_and_split()
```

```python
bm2.fit_evaluate("first_time")
```

```python
bm2.plot_feature_importances("first_time")
```

```python
bm2.shap_values("first_time", summary=True, bar=True, force=False)
```

```python
# First-time cohort, shallow tree for readability
bm2.plot_sample_tree("first_time", estimator_idx=0, max_depth=3, save_path="first_time_tree")
```

```python
bm2.fit_evaluate("previously_contacted")
```

```python
bm2.plot_feature_importances("previously_contacted")
```

```python
bm2.shap_values("previously_contacted", summary=True, bar=True, force=False)
```

```python
bm2.plot_sample_tree("previously_contacted", estimator_idx=5, max_depth=3, save_path="previously_contacted_tree")
```
